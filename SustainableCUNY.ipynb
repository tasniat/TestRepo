# **Phase 2: Excel Conversions**

```python
from datetime import datetime
```

```python
# Created an index column to later use as the primary key for SQL tables
dfnyserda['id'] = dfnyserda.index + 1
dfnyserda.head()
```

```python
# Unnecessary column
dfnyserda.drop(['legacy_project_number'], inplace=True, axis=1)
dfnyserda.head()
```

```python
# Converted data to desired format of YY-MM-DD HH:MM:SS AM
d = ''
dt = []
for x in dfnyserda['date_application_received']: 
  dpre = x.split('T')
  dprf = dpre[1].split('.')
  d = dpre[0] + '-' + dprf[0]
  d = datetime.strptime(d, "%Y-%m-%d-%H:%M:%S")
  d = d.strftime("%Y-%m-%d %H:%M:%S AM")
  dt.append(d)
```

```python
dfnyserda['date_application_received'] = dt
```

```python
dfnyserda['date_application_received'].head(100)
```

```python
d = ''
dt = []

from numpy import NaN
for x in dfnyserda['date_install']: 
  if(str(x) != NaN):
    dpre = str(x).split('T')
    if(dpre[0] != 'nan'): # This was the only column with dates that had blanks in some rows so I had to account for that
      dprf = dpre[1].split('.')
      d = dpre[0] + '-' + dprf[0]
      d = datetime.strptime(d, "%Y-%m-%d-%H:%M:%S")
      d = d.strftime("%Y-%m-%d %H:%M:%S AM")
      dt.append(d)
    else:
      dt.append(' ')
```

```python
dfnyserda['date_install'] = dt
dfnyserda['date_install'].head(50)
```

```python
d = ''
dt = []
for x in dfnyserda['reporting_period']: 
  dpre = x.split('T')
  dprf = dpre[1].split('.')
  d = dpre[0] + '-' + dprf[0]
  d = datetime.strptime(d, "%Y-%m-%d-%H:%M:%S")
  d = d.strftime("%Y-%m-%d %H:%M:%S AM")
  dt.append(d)
```

```python
dfnyserda['reporting_period'] = dt
```
